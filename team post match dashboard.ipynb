```python

def get_passes_df(events_dict):
    data = pd.DataFrame(events_dict)
    # create receiver column based on the next event
    # this will be correct only for successfull passes
    data["pass_recipient"] = data["playerName"].shift(-1)
    # filter only passes

    passes_ids = data.index[data['event_type'] == 'Pass']
    data_passes = data.loc[
        passes_ids, ["id", "minute", "x", "y", "endX", "endY", "teamId", "playerId", "playerName", "event_type",
                     "outcomeType", "pass_recipient", "isTouch", 'Venue']]

    return data_passes


```

```python
def get_player_dfs(data):
    # Create an empty dictionary to store dataframes for each player
    player_dfs = {}

    # Loop over the unique values in the 'playerName' column
    for player_name in data['playerName'].dropna().unique():

        # Create a new dataframe for the player
        player_df = data[data['playerName'] == player_name]

        # Get the average x and y values for the player
        x_avg = player_df['x'].mean()
        y_avg = player_df['y'].mean()

        # Add the x and y averages to the dataframe
        player_df['x_avg'] = x_avg
        player_df['y_avg'] = y_avg

        # Get the player's initials
        initials = player_name.split()[0][0] + player_name.split()[-1][0]

        # Add the initials to the dataframe
        player_df['initials'] = initials

        # Add the dataframe to the dictionary with the player name as the key
        player_dfs[player_name] = player_df

    return player_dfs
```
