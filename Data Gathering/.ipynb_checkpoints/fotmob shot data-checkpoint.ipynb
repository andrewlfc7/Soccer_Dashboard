```python
match_id = feb_fix.id[1]
response = requests.get(f'https://www.fotmob.com/api/matchDetails?matchId={match_id}&ccode3=USA&timezone=America%2FChicago&refresh=true&includeBuzzTab=false&acceptLanguage=en-US')
data = response.content
data = json.loads(data)
matchId = data['general']['matchId']

leagueId = data['general']['leagueId']
leagueId

team_logos = data['header']['teams']
team_logos = pd.DataFrame(team_logos)
team_logos


homeTeam = data['general']['homeTeam']
awayTeam = data['general']['awayTeam']

homeTeam = pd.DataFrame(homeTeam,index=[0])
awayTeam = pd.DataFrame(awayTeam,index=[0])

shot_data = data['content']['shotmap']['shots']

general = data ['general']
general = pd.DataFrame(general)
general

df_shot = pd.DataFrame(shot_data)

df_shot['matchId'] = matchId
df_shot

def extract_value(d, key):
    return d[key]


df_shot['x_value'] = df_shot['onGoalShot'].apply(extract_value, args=('x',))
df_shot['y_value'] = df_shot['onGoalShot'].apply(extract_value, args=('y',))
df_shot['zoomRatio_value'] = df_shot['onGoalShot'].apply(extract_value, args=('zoomRatio',))
df_shot.drop(['onGoalShot'], axis=1, inplace=True)


df_shot = df_shot.rename(columns={'x_value':'onGoalShot_X'})
df_shot = df_shot.rename(columns={'y_value':'onGoalShot_Y'})
df_shot = df_shot.rename(columns={'zoomRatio_value':'onGoalShot_ZR'})

df_shot.to_sql("Matches", con, if_exists="append")

```
